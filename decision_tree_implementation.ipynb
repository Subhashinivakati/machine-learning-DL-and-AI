```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt

# Load dataset
iris = load_iris()
X = iris.data  # Features
y = iris.target  # Target classes (0: setosa, 1: versicolor, 2: virginica)
feature_names = iris.feature_names
class_names = iris.target_names  # Class labels

# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Full Tree (No Pruning)
full_tree = DecisionTreeClassifier(random_state=42)
full_tree.fit(X_train, y_train)

# Plot the full tree
plt.figure(figsize=(20, 10))
plot_tree(full_tree, feature_names=feature_names, class_names=class_names, filled=True)
plt.title("Full Decision Tree (No Pruning)")
plt.show()
```

```python
# Evaluate
y_pred_full = full_tree.predict(X_test)
print("Full Tree Results:")
print(f"Test Accuracy: {accuracy_score(y_test, y_pred_full):.2f}")
print(classification_report(y_test, y_pred_full, target_names=class_names))
```

## pre-prunning

```python
pre_pruned_tree = DecisionTreeClassifier(
    max_depth=3, 
    min_samples_split=5, 
    random_state=42
)
pre_pruned_tree.fit(X_train, y_train)
```

```python
# Plot pre-pruned tree
plt.figure(figsize=(20, 10))
plot_tree(pre_pruned_tree, feature_names=feature_names, class_names=class_names, filled=True)
plt.title("Pre-Pruned Decision Tree (max_depth=3)")
plt.show()
```

```python
# Evaluate
y_pred_pre = pre_pruned_tree.predict(X_test)
print("\nPre-Pruned Tree Results:")
print(f"Test Accuracy: {accuracy_score(y_test, y_pred_pre):.2f}")
print(classification_report(y_test, y_pred_pre, target_names=class_names))
```
