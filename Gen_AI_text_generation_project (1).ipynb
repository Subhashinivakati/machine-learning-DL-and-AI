# Step 2: Import the 'pipeline' function and load the GPT-2 model

```python
# The 'pipeline' function simplifies using pre-trained models for various tasks.
from transformers import pipeline
```

```python
In Hugging Face Transformers, a high-level API refers to the pipeline() function,
a user-friendly interface designed to simplify the usage of pre-trained models for common NLP and other tasks.
```

```python
# Load the text generation pipeline using the 'gpt2' model.
# This downloads the pre-trained GPT-2 model weights.
generator = pipeline('text-generation', model='gpt2')
print("GPT-2 model loaded successfully! Ready to generate text.")
```

```python
# Text Generation Parameters

# 1.max_length-
This parameter specifies the maximum total number of tokens (words or sub-word units) the generated text should have,
including your initial input prompt.

For example, if your prompt is 10 tokens long and max_length is 50, the model will generate up to 40 new tokens.

# 2.num_return_sequences-
This parameter determines how many independent and different sequences of text the model
should generate for the given input prompt and parameters.
- You get multiple distinct outputs for the same prompt.

# 3.temperature- 0 to 1
This parameter controls the randomness or "creativity" of the generated text.


# 4. no_repeat_ngram_size
This parameter prevents the model from generating n-grams (sequences of 'n' words/tokens) that have already appeared in the generated text.
prevents the model from repeating short phrases.
Generally improves output quality without a significant performance hit.


# 5.top_k
This parameter limits the set of possible next words to the top k most probable words at each step of generation.

High top_k ((e.g., 100 or more))- Allows the model to choose from a wider range of words.
(Can lead to more diverse and sometimes more creative outputs, but also increases the chance of generating less coherent or off-topic text,
 as it includes more "risky" words in the selection pool.)

Low top_k (e.g., 10 or 20)-
(Restricts the model's choices to only the very most probable words.)
Makes the output more focused and predictable, but can also make it generic, bland, or lead to repetition

# 6.top_p (Nucleus Sampling)-
(This parameter selects the smallest set of most probable words whose cumulative probability exceeds p. The model then samples from only these words.)


## "best" values often depend on the specific model, the prompt, and the desired outcome!
```

# Step 3: Generate text with a simple prompt

```python
# Define our first prompt
my_prompt = "Once upon a time, in a land far, far away, there lived a brave knight who"
```

```python
# Generate text using our generator
# max_length: The maximum number of words (tokens) the generated text should have.
# num_return_sequences: How many different outputs we want for the same prompt.
# We'll set it to 1 for simplicity first.


generated_text = generator(my_prompt, max_length=50, num_return_sequences=1)
```

```python
# Print the generated text
print("--- Generated Text 1 ---")
print(generated_text[0]['generated_text'])
print("-" * 25)
```

# Step 5: Explore more parameters - 'temperature'

```python
# A prompt for a more creative output
creative_prompt = "Write a short poem about a lost star in the night sky."

# Generate text with 'temperature' parameter
# temperature: Controls the randomness of the output.
# A higher temperature (e.g., 0.7 to 1.0) makes the output more random/creative.
# A lower temperature (e.g., 0.1 to 0.5) makes the output more focused/less random.
# Default is usually 1.0 or similar.
generated_text_3 = generator(creative_prompt, max_length=70, num_return_sequences=1, temperature=0.9)

print("\n--- Generated Text 3 (High Temperature) ---")
print(generated_text_3[0]['generated_text'])


generated_text_4 = generator(creative_prompt, max_length=70, num_return_sequences=1, temperature=0.3)

print("\n--- Generated Text 4 (Low Temperature) ---")
print(generated_text_4[0]['generated_text'])
print("-" * 25)
```

```python
# Exploring advanced parameters for better output

# A prompt for a more creative output
creative_prompt = "Write a short poem about a lost star in the night sky."

# Generate text with 'temperature' and 'no_repeat_ngram_size'
# no_repeat_ngram_size=2 will prevent repeating sequences of 2 words (bigrams).
# we can also try top_k or top_p for more nuanced control.
generated_text_3_improved = generator(
    creative_prompt,
    max_length=100,
    num_return_sequences=1,
    temperature=0.8,
    no_repeat_ngram_size=2,
    # top_k=50,
    # top_p=0.9
)

print("\n--- Generated Text 3 (High Temperature + No Repeat N-gram) ---")
print(generated_text_3_improved[0]['generated_text'])

# Try with a lower temperature for comparison, still using no_repeat_ngram_size
generated_text_4_improved = generator(
    creative_prompt,
    max_length=100,
    num_return_sequences=1,
    temperature=0.5,
    no_repeat_ngram_size=2
)


print("-*" * 25)


print("\n--- Generated Text 4 (Low Temperature + No Repeat N-gram) ---")
print(generated_text_4_improved[0]['generated_text'])
print("-*" * 25)
```

```python
prompt_1 = "Write a short story about a child's first Diwali celebration, filled with lights, sweets, and family joy in a small Indian village."

generated_indian_text_1 = generator(
    prompt_1,
    max_length=120, # Give it enough length to develop
    num_return_sequences=1,
    temperature=0.8,
    no_repeat_ngram_size=2
)
print("\n--- Diwali Story ---")
print(generated_indian_text_1[0]['generated_text'])
print("-" * 25)
```

```python


# Refined Prompt for a more guided story
# We're making it more specific: character name, age, what they see, feel, and do.
refined_diwali_prompt = """
Write a heartwarming short story about a 7-year-old girl named Meera experiencing her very first grand Diwali celebration in her bustling Indian neighborhood.
Focus on the vibrant sights of flickering diyas, the sweet aroma of ladoos, the joyous sounds of firecrackers, and Meera's innocent wonder.
The story should capture the spirit of family, togetherness, and the magic of the festival.
"""

# Generate text with refined parameters
# We'll use a combination of temperature, top_k, and top_p for better control
# and definitely no_repeat_ngram_size to avoid repetition.
generated_diwali_story = generator(
    refined_diwali_prompt,
    max_length=250,
    num_return_sequences=1,
    temperature=0.9,
    no_repeat_ngram_size=3,
    top_k=50,
    top_p=0.9,

)

print("\n--- Refined Diwali Story ---")
# Access the generated text from the first sequence in the list of dictionaries
print(generated_diwali_story[0]['generated_text'])
print("-" * 25)
```

```python
# Updated Code Cell for "Diwali Story"

# Hyper-Specific Prompt for a focused story segment
# We're giving it a very direct opening and clear instructions for the content.
refined_diwali_prompt_2 = """
Story: Meera's First Diwali

Meera, a curious 7-year-old, peered out from her grandmother's arms. It was her first grand Diwali celebration in their bustling Mumbai chawl. The air hummed with excitement.

Describe:
- The dazzling rows of flickering diyas on every windowsill.
- The sweet, intoxicating smell of fresh jalebis and ladoos from the kitchen.
- The distant, joyous sounds of children laughing and small firecrackers popping.
- Meera's wide-eyed wonder and pure happiness as she soaked in the festive atmosphere.
- A brief mention of family members smiling around her.

Keep the tone heartwarming and focused on sensory details.
"""

# Generate text with aggressive parameter tuning
generated_diwali_story_2 = generator(
    refined_diwali_prompt_2,
    max_length=150,
    num_return_sequences=1,
    temperature=0.6,
    no_repeat_ngram_size=3,
    top_k=50,
    top_p=0.9,
)

print("\n--- Refined Diwali Story (Attempt 3) ---")
print(generated_diwali_story_2[0]['generated_text'])
print("-" * 25)
```
