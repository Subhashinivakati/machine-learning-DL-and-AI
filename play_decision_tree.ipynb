```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

# Step 1: Create DataFrame
data = {
    "Outlook": ["Sunny", "Sunny", "Rainy", "Rainy", "Sunny"],
    "Temperature": ["Hot", "Hot", "Cold", "Cold", "Cold"],
    "Wind": ["Strong", "Weak", "Strong", "Weak", "Weak"],
    "Play": ["No", "No", "Yes", "No", "Yes"]
}
df = pd.DataFrame(data)

# Step 2: Encode categorical variables
le_outlook = LabelEncoder()
df["Outlook"] = le_outlook.fit_transform(df["Outlook"])
df["Temperature"] = LabelEncoder().fit_transform(df["Temperature"])
df["Wind"] = LabelEncoder().fit_transform(df["Wind"])
df["Play"] = LabelEncoder().fit_transform(df["Play"])

# Step 3: Train the model
X = df[["Outlook", "Temperature", "Wind"]]
y = df["Play"]
clf = DecisionTreeClassifier()
clf.fit(X, y)

# Step 4: Visualize the tree
plt.figure(figsize=(10,6))
plot_tree(clf, feature_names=["Outlook", "Temperature", "Wind"], 
          class_names=["No", "Yes"], filled=True)
plt.show()
```
